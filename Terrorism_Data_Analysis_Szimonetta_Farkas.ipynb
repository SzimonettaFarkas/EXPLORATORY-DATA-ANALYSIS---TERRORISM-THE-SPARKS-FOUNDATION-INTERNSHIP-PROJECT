```python
# Creating a Copy of the Dataset
data1 = data.copy()
data1.head()
```

```python
# Data Preprocessing, Cleaning
# Looking for Missing Values
data1.isna().sum()
```

```python
# Percentage of Missing Values
null_sum = (data1.isnull().sum()/len(data))*100
null_sum
```

```python
# Sorting Columns According to the Percentage of Missing Values
percent_null=null_sum.sort_values(ascending=False)
percent_null
```

```python
# Columns with more than 50% Mising Values
more_than_50 = percent_null[percent_null>=50]
more_than_50
```

```python
# Columns with less than 50% Missing Values
less_than_50=percent_null[~percent_null.isin(more_than_50)]
less_null=less_than_50.index.to_list()
print(len(less_null))
print(less_null)
```

```python
dataset = data1[less_null]
dataset.head()
```

```python
checking = dataset.columns.to_list()
selected_columns=['eventid', 'iyear', 'imonth', 'iday','country_txt','region_txt','city','multiple', 'success', 
                    'suicide', 'attacktype1_txt', 'targtype1_txt',  'targsubtype1_txt', 'corp1', 
                    'target1', 'natlty1_txt', 'weaptype1_txt','nkill','nwoundus', 'claimed', 'individual', 'INT_LOG', 
                    'doubtterr', 'INT_MISC', 'specificity','gname', 'ishostkid', 'INT_ANY', 'guncertain1', 'provstate']
```

```python
# Exploratory Data Analysis
# Attack per Year
attack_per_year=df['iyear'].value_counts()
ax=plt.rcParams['figure.figsize'] = 22,8.27
ax=sns.countplot(x='iyear',data=df, color = 'blue')
ax.set_xlabel('Year')
plt.xticks(rotation = 45)
ax.set_ylabel('Number of Terrorist Attack')
ax.set_title('Number of Terrorist Attacks Per Year')
```

```python
# Top 15 Most Attacked Countries
ax=plt.rcParams['figure.figsize'] = 22,8.27
ax=sns.countplot(x='country_txt',data=df,order = df['country_txt'].value_counts().index[:15], color  = 'blue')
ax.set_xlabel('Country')
plt.xticks(rotation = 75)
ax.set_ylabel('Number of Terrorist Attacks')
ax.set_title('Top 15 Most Attacked Countries')
```

```python
# Number of Attacks per Region
ax=plt.rcParams['figure.figsize'] = 16,8.27
ax=sns.countplot(x='region_txt',data=df,order = df['region_txt'].value_counts().index, color = 'green')
ax.set_xlabel('Region',fontsize=18)
plt.xticks(rotation = 75)
ax.set_ylabel('Number of Terrorist Attacks')
ax.set_title('Number of Attacks per Region')
```

```python
# Top 5 Most Frequently Used Weapon Types
ax=plt.rcParams['figure.figsize'] = 16,8.27
ax=sns.countplot(x='weaptype1_txt',data=df,order = df['weaptype1_txt'].value_counts().index[:5], color = 'green')
ax.set_xlabel('Weapon Type')
plt.xticks(rotation = 75)
ax.set_ylabel('Number of Attacks')
ax.set_title('Top 5 Most Frequently Used Weapon Types')
```

```python
# Top 5 Most Frequent Type of Attacks
ax=plt.rcParams['figure.figsize'] = 16,8.27
ax=sns.countplot(x='attacktype1_txt',data=df,order = df['attacktype1_txt'].value_counts().index[:5], color = 'red')
ax.set_xlabel('Type of Attack')
plt.xticks(rotation = 75)
ax.set_ylabel('Number of Attacks')
ax.set_title('Top 5 Most Frequent Type of Attacks')
```

```python
# Top 10 Most Frequents Targets
ax=plt.rcParams['figure.figsize'] = 16,8.27
ax=sns.countplot(x='targtype1_txt',data=df,order = df['targtype1_txt'].value_counts().index[:10], color = 'red')
ax.set_xlabel('Type of Targets')
plt.xticks(rotation = 75)
ax.set_ylabel('Number of Attacks')
ax.set_title('Top 10 Most Frequents Targets')
```

```python
# Top 15 Terrorist Organizations Made Most Number of Attacks
ax=plt.rcParams['figure.figsize'] = 16,12
ax=sns.countplot(y='gname',data=df,order = df['gname'].value_counts().index[1:15], color = 'black')
ax.set_ylabel('Terrorist Organizations')
# plt.yticks(rotation = 75)
ax.set_xlabel('Number of Attacks')
ax.set_title('Top 15 Terrorist Organizations Made Most Number of Attacks')
```

```python
# Ratio of Successful Terrorist Attacks
colors = ['red', 'green']
ax=df['success'].value_counts().plot(kind='pie', colors = colors)
ax.set_title('Ratio of Successful Terrorist Attacks')
```

```python
# Performance of Countries Against Terrorist Attacks
# Top 10 Countries with the Best Performance Against Terrorist Attacks (More than 5 Attacks)
success_country=df.groupby('country_txt').agg({'eventid':'count','success':'sum'}).reset_index()
success_country.columns=['Country','Total_Attacks','Attack_Success']
success_country['Attack_Failed']=success_country['Total_Attacks']-success_country['Attack_Success']
success_country['Failure_Rate_Attack']=(success_country['Attack_Failed']/success_country['Total_Attacks'])*100
success_country['Success_Rate_Attack']=(success_country['Attack_Success']/success_country['Total_Attacks'])*100
success_country=success_country.round(2)

able_stop_terror_attack=success_country[success_country['Total_Attacks']>=5].sort_values('Failure_Rate_Attack',
                                                                                         ascending=False )[:10]
plot_able_stop_terror_attack=able_stop_terror_attack.loc[:,['Country','Failure_Rate_Attack']]
ax=plt.rcParams['figure.figsize'] = 16,8.27
ax=sns.barplot(x='Country',y='Failure_Rate_Attack',data=plot_able_stop_terror_attack, color = 'black')
ax.set_xlabel('Countries')
plt.xticks(rotation = 75)
ax.set_ylabel('Ratio of Failure of Terror Attacks')
ax.set_title("Top 10 Countries with the Best Performance Against Terrorist Attacks")
```

```python
# Top 10 Countries with the Worst Performance Against Terrorist Attacks (More than 5 Attacks)
unable_stop_terror_attack=success_country[success_country['Total_Attacks']>=5].sort_values('Success_Rate_Attack'
                                                                                           ,ascending=False )[:10]
plot_unable_stop_terror_attack=unable_stop_terror_attack.loc[:,['Country','Success_Rate_Attack','Total_Attacks']]
ax=plt.rcParams['figure.figsize'] = 16,8.27
ax=sns.barplot(x='Country',y='Success_Rate_Attack',data=plot_unable_stop_terror_attack, color = 'red')
ax.set_xlabel('Countries')
plt.xticks(rotation = 75)
ax.set_ylabel('Ratio of Success of Terror Attack')
ax.set_title('Top 10 Countries with the Worst Performance Against Terrorist Attacks ')
```

```python
# Terrorist Attack Trends
# Weapon Type Trends
attack_per_year=df.groupby(['iyear','targtype1_txt','target1','success','weaptype1_txt',
                              'gname'])['eventid'].count().to_frame(name = 'Total_Attacks').reset_index()
ax = sns.relplot(col="weaptype1_txt", y="Total_Attacks",col_wrap=4, hue='weaptype1_txt',x="iyear"
                 ,kind='line', ci=None ,data=attack_per_year, height=3, aspect=1.5, legend=False)
```

```python
# Target Type Trends
ax = sns.relplot(col="targtype1_txt", y="Total_Attacks",col_wrap=5,
                 hue='targtype1_txt',x="iyear",kind='line', ci=None ,data=attack_per_year, height=3, aspect=1.5, legend=False)
```

```python
# Trend of Successful/Unsuccessful Terrorist Attacks
ax = sns.lineplot(x="iyear", y="Total_Attacks", hue='success', ci=None ,data=attack_per_year)
ax.tick_params(rotation=45)
```
